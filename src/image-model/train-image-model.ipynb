# Training the YOLO Classify Model
This code trains the yolo model on the frames extracted from the data set. The images must be placed into the test and train subfolders in the subfolder that corresponds with the emotion label.

```python
from ultralytics import YOLO

# Load the YOLO-pose nano model, pose version
model = YOLO('yolov8m-cls')

# Train the model with our data set defined in dogs-pose.yaml
results = model.train(data='dataset/', epochs=50, imgsz=120)
```

### Training Summary.
The model train for 50 epochs over 3.22 hours. The accuracy is 48.9%
