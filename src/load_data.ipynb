```python
import cv2
import os

def save_frames(video_path, video_number, label):
    # This function will extract 1 frame per second from a video and same
    # The images to the frame folder in a subfolder the same name as the label

    # Load video
    stream = cv2.VideoCapture(video_path)
    # Initialize frame number
    frame_number = 0

    # Loop through every frame in the video stream
    while(stream.isOpened()):
        # Get return_value and frame from the stream
        ret, frame = stream.read()
        if ret:
            # Only save 1 frame per second
            if (frame_number % 24 == 0):
                # Convert image to gray scale
                #gray_image = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                # Resize image to be 120x120
                image = cv2.resize(frame, (120,120))

                # Save image
                file_name = 'data/frames/' + label + '/' + str(video_number) + '_' + str(frame_number / 24) + '.jpg'
                cv2.imwrite(file_name, image)
                frame_number += 1
            frame_number += 1
        else:
            break

    # Ensure all streams are released and closed properly
    stream.release()
    cv2.destroyAllWindows()
```

```python
import pandas as pd

# Read data.csv
data = pd.read_csv('labels.csv')

# Read video paths to a list
videos = data['path'].tolist()
video_labels = data['video_label'].tolist()

# Extract frames from the video clips

video_number = 0

for clip in videos:
  label = video_labels[video_number]
  save_frames(clip, video_number, label)

  video_number += 1



```
